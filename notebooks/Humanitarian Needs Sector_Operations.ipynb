The Integration of Humanitarian Needs Data by Sector by Region fpor 2018, 2019 and 2020

```python
#import lib
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#READ 2018 data only the people in need per region and drop the second line
df_pin_2018= pd.read_excel('ethiopia-hno-2018_hxl.xlsx', sheet_name='People in need Per region', header=[0], index_col='Region')
#df_pin_2018.reset_index()
print('2018 People in Need Data')
print(df_pin_2018.info())
```

```python
print('2018 People in Need Data Description')
print(df_pin_2018.describe())
```

```python
#drop the first row, make all columns lower case and rename the columns to make them same as 2019, 2020 data
df_pin_2018=df_pin_2018.drop(df_pin_2018.index[0])
df_pin_2018.drop(df_pin_2018.tail(1).index,inplace=True) # drop last row
df_pin_2018.index.names = ['ADM1_EN']
```

```python
df_pin_2018.columns =df_pin_2018.columns.str.lower()
df_pin_2018 = df_pin_2018.drop(df_pin_2018.columns[-1],axis=1)
df_pin_2018 = df_pin_2018.drop(df_pin_2018.columns[-1],axis=1)
df_pin_2018['Year']='2018'
df_pin_2018
```

```python
#READ 2019 data only the people in need per region and drop the second line make all columns lower case and rename the columns to make them same as 2018, 2020 data
df_pin_2019= pd.read_excel('ethiopia-humanitarian-needs-overview_2019.xlsx', sheet_name='Needs by Sector', header=[0], index_col='REGION')
print('2019 People in Need Data')
print(df_pin_2019.info())
```

```python
print('2019 People in Need Data Description')
print(df_pin_2019.describe())
```

```python
df_pin_2019.rename(columns={'SHELTER & NFI':'ESNFI','FOOD INSECURITY':'food','AGRICULTURE AND LIVESTOCK':'Agriculture'}, inplace=True)
df_pin_2019.index.names = ['ADM1_EN']
df_pin_2019.columns =df_pin_2019.columns.str.lower()
#df_pin_2018.reset_index()
df_pin_2019.head()
```

```python
#drop the first row 
df_pin_2019=df_pin_2019.drop(df_pin_2019.index[0])
df_pin_2019['Year']='2019'
```

```python
#READ 2020 data only the people in need per region and drop the second line
df_pin_2020= pd.read_excel('ethiopia-2020-humanitarian-needs-overview.xlsx', sheet_name='Sectoral PIN by Admin3', header=[0], index_col='admin1Name_en')
#df_pin_2018.reset_index()
print('2020 People in Need Data')
print(df_pin_2019.info())
```

```python
print('2020 People in Need Data Description')
print(df_pin_2020.describe())
```

```python
#drop the first row 
df_pin_2020=df_pin_2020.drop(df_pin_2020.index[0])
df_pin_2020.index.names = ['ADM1_EN']
df_pin_2020.columns =df_pin_2020.columns.str.lower()
df_pin_2020.head()
```

```python
#agregate by region to be similar to 2018 and 2019
df_pin_2020_region=(df_pin_2020.groupby("ADM1_EN", as_index=True)
           .agg({"agriculture":"sum", "education":"sum", "esnfi":"sum", "health":"sum", "food":"sum", "nutrition":"sum", "protection":"sum", "wash":"sum"}))
```

```python
#add year
df_pin_2020_region['Year']='2020'
df_pin_2020_region.head()
```

```python
#combining sector data for 2018,2019 and 2020 by region #adm1+nameADM1_EN
frames = [df_pin_2018, df_pin_2019, df_pin_2020]
df_pin_years=pd.concat(frames, sort=False)
df_pin_years.head()
df_pin_years.to_csv('humanitarian_needs_by_region_2018_19_20')
```

Integration of Humanitarian Needs by Sector By Woreda and Operations 2020

```python
#add dataframe who is doing what per Woreda Name
df_2020_op= pd.read_excel('ethiopia-who-is-doing-what-where-3w-august-2020.xlsx', sheet_name='Programme&Implementing part', header=[0])
#df_pin_2018.reset_index()
#drop the first row 
df_2020_op=df_2020_op.drop(df_2020_op.index[0])
print('2020 Operations data')
print(df_2020_op.info())
```

```python
#rename column for indexing
df_2020_op.rename(columns={'Region':'ADM1_EN','Zone':'ADM2_EN','Woreda':'ADM3_EN'}, inplace=True)
#df_2020_op = df_2020_op.set_index(['ADM1_EN','ADM2_EN','ADM3_EN'])
```

```python
#READ 2020 data only the people in need per adm3_ENG 
df_pin_2020= pd.read_excel('ethiopia-2020-humanitarian-needs-overview.xlsx', sheet_name='Sectoral PIN by Admin3', header=[0])
#df_pin_2018.reset_index()
print('2020 People in Need Data')
print(df_pin_2019.info())
```

```python
#drop the first row 
df_pin_2020=df_pin_2020.drop(df_pin_2020.index[0])
#df_pin_2020.index.names = ['ADM3_EN']
#change names based on defined keys
df_pin_2020.rename(columns={'admin1Name_en':'ADM1_EN','admin2Name_en':'ADM2_EN', 'admin3Name_en':'ADM3_EN'}, inplace=True)
```

```python
#drop ids
df_pin_2020.drop(columns=['admin1Pcode','admin2Pcode','admin3Pcode'],axis=1, inplace=True)
```

```python
df_pin_2020 = pd.melt(df_pin_2020, id_vars=["ADM1_EN","ADM2_EN","ADM3_EN"])
#df_pin_2020 = df_pin_2020.set_index(['ADM1_EN','ADM2_EN','ADM3_EN'])
df_pin_2020.rename(columns={'variable':'Sector','value':'#PIN'}, inplace=True)


#df_pin_2020.melt(id_vars=['ADM3_EN'],var_name="Sector", value_name="PIN")
```

```python
#add 2020 sector data with who id doing what
df_pin_op_2020=pd.merge(df_pin_2020, df_2020_op, on=['ADM1_EN','ADM2_EN','ADM3_EN','Sector'], how='outer').fillna("none")
#df_pin_op_2020 =df_pin_op_2020.set_index(['ADM3_EN'])
df_pin_op_2020.to_csv('humanitarian_needs_and operations_by_woreda_2020')
```
